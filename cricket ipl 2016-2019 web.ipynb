```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
import re
import time
```

```python
players=[]
innings=[]
notout=[]
runs=[]
highscore=[]
average=[]
balls_faced=[]
strike_rate=[]
hundreds=[]
fifty=[]
four=[]
six=[]
year=[]
```

```python
for x in range(2015,2020):
    URL = 'https://www.iplt20.com/stats/{}/most-runs'.format(x)
    page=requests.get(URL)
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)

    for i in soup.find_all("div",attrs={"class":"top-players__player-name"}):
        if players is None:
            player.append(np.NaN)
        else:
            players.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__inns"}):
        if innings is None:
            innings.append(np.NaN)
        else:
            innings.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__no"}):
        if notout is None:
            notout.append(np.NaN)
        else:
            notout.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__r"}):
        if runs is None:
            runs.append(np.NaN)
        else:
            runs.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__hs"}):
        if highscore is None:
            highscore,append(np.NaN)
        else:
            highscore.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__a"}):
        if average is None:
            average.append(np.NaN)
        else:
            average.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__b"}):
        if balls_faced is None:
            balls_faced.append(np.NaN)
        else:
            balls_faced.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__sr"}):
        if strike_rate is None:
            strike_rate.append(np.NaN)
        else:
            strike_rate.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__100s"}):
        if hundreds is None:
            hundreds.append(np.NaN)
        else:
            hundreds.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__50s"}):
        if fifty is None:
            fifty.append(np.NaN)
        else:
            fifty.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__4s"}):
        if four is None:
            four.append(np.NaN)
        else:
            four.append(i.text)
    for i in soup.find_all("td",attrs={"class":"top-players__6s"}):
        if six is None:
            six.append(np.NaN)
        else:
            six.append(i.text)
    for i in range(0,100):
        year.append(x)
                        
```

```python
IPL= pd.DataFrame({'Players':players,'Innings':innings,'Notout':notout,'Runs':runs,'High_Score':highscore,'Average':average,'Balls_Faced':balls_faced,'Strike_rate':strike_rate,'100_s':hundreds,'50_s':fifty,'4_s':four,'6_s':six,'Years':year})
```

```python
regex='[a-zA-Z]+'
IPL['Players']=IPL['Players'].apply(lambda x:re.compile(regex).findall(x))
```

```python
regex='\d+'
IPL['Innings']=IPL['Innings'].apply(lambda x:re.compile(regex).findall(x))
IPL['Notout']=IPL['Notout'].apply(lambda x:re.compile(regex).findall(x))
IPL['Runs']=IPL['Runs'].apply(lambda x:re.compile(regex).findall(x))
IPL['High_Score']=IPL['High_Score'].apply(lambda x:re.compile(regex).findall(x))
IPL['4_s']=IPL['4_s'].apply(lambda x:re.compile(regex).findall(x))
IPL['Balls_Faced']=IPL['Balls_Faced'].apply(lambda x:re.compile(regex).findall(x))
IPL['100_s']=IPL['100_s'].apply(lambda x:re.compile(regex).findall(x))
IPL['50_s']=IPL['50_s'].apply(lambda x:re.compile(regex).findall(x))
IPL['6_s']=IPL['6_s'].apply(lambda x:re.compile(regex).findall(x))
```

```python
regex='\d+\.+\d+'
IPL['Average']=IPL['Average'].apply(lambda x:re.compile(regex).findall(x))
IPL['Strike_rate']=IPL['Strike_rate'].apply(lambda x:re.compile(regex).findall(x))
```

```python
IPL['Players']=IPL['Players'].apply(lambda x:''.join(x))
IPL['Innings']=IPL['Innings'].apply(lambda x:''.join(x)).astype(int)
IPL['Notout']=IPL['Notout'].apply(lambda x:''.join(x)).astype(int)
IPL['Runs']=IPL['Runs'].apply(lambda x:''.join(x)).astype(int)
IPL['High_Score']=IPL['High_Score'].apply(lambda x:''.join(x)).astype(int)
IPL['Average']=IPL['Average'].apply(lambda x:''.join(x))
IPL['Balls_Faced']=IPL['Balls_Faced'].apply(lambda x:''.join(x)).astype(int)
IPL['Strike_rate']=IPL['Strike_rate'].apply(lambda x:''.join(x)).astype(float)
IPL['100_s']=IPL['100_s'].apply(lambda x:''.join(x)).astype(int)
IPL['50_s']=IPL['50_s'].apply(lambda x:''.join(x)).astype(int)
IPL['4_s']=IPL['4_s'].apply(lambda x:''.join(x)).astype(int)
IPL['6_s']=IPL['6_s'].apply(lambda x:''.join(x)).astype(int)
```
